```python
import pandas as pd
```

```python
df = pd.read_csv('dd.csv', index_col=[0])
```

```python
df.tail(50)
```

```python
from lifelines import KaplanMeierFitter
```

```python
kmf = KaplanMeierFitter()
```

```python
T = df['duration']
C = df['observed']
```

```python
kmf.fit(T,C)
```

```python
%pylab inline
figsize(12,6)
```

```python
kmf.plot()
```

```python
df['regime'].unique()
```

```python
ax = subplot(111)

for r in df['un_continent_name'].unique()[:]:
    ix = df['un_continent_name'] == r
    kmf.fit(T.ix[ix], C.ix[ix], label=r)
    ax = kmf.plot(ax=ax)
```
